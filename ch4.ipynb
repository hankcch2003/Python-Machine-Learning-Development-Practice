# 3. Rolling

```python
import pandas as pd
df = pd.read_csv("starbucks.csv", index_col="Date", parse_dates = True)
df.index
```

```python
df.head()
```

```python
df["Close"].plot(figsize=(12,5))
```

```python
newDf = df.rolling(window = 7).mean().head(15)
newDf = newDf.bfill()
newDf
```

```python
newDf.iloc[0:7] = 0
newDf
```

```python
df["Close"].plot(figsize=(12,5))
df.rolling(window = 30).mean()["Close"].plot(figsize=(12,5))
```

```python
df["Close"].plot(figsize=(12,5))
df.rolling(window = 5).mean().bfill()["Close"].plot(figsize=(12,5))
df.rolling(window = 15).mean().bfill()["Close"].plot(figsize=(12,5))
df.rolling(window = 30).mean().bfill()["Close"].plot(figsize=(12,5))
```
