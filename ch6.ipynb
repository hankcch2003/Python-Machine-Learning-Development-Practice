```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

# 5. Inner, Left, Right, and Outer Joins

```python
pd.merge(registrations, logins, how="inner", on="name")
```

```python
pd.merge(registrations, logins, on="name")
```

```python
pd.merge(registrations, logins, how="left", on="name")
```

```python
pd.merge(registrations, logins, how="right", on="name")
```

```python
pd.merge(registrations, logins, how="outer", on="name")
```

# 6. Pivot Tables(æ¨žç´åˆ†æžè¡¨)

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Sales_Funnel_CRM.csv")
df
```

```python
license = df[["Company","Product","Licenses"]]
license
```

```python
pd.pivot(data = license, index="Company", columns="Product", values="Licenses")
```

```python
df
```

```python
myData = pd.pivot_table(df, index="Company", aggfunc="sum")
print(type(myData))
myData
```

```python
myData[["Licenses","Sale Price"]]
```

```python
pd.pivot_table(df, index="Company", aggfunc="sum", values=["Licenses","Sale Price"])
```

```python
pd.pivot_table(df, index=["Account Manager","Contact"], values=["Sale Price","Licenses"], aggfunc="sum")
```

```python
pd.pivot_table(df, index=["Account Manager","Contact"], values=["Sale Price","Licenses"], columns=["Product"], aggfunc="sum")
```

```python
pd.pivot_table(df, index=["Account Manager","Contact"], values=["Sale Price","Licenses"], columns=["Product"], aggfunc=["sum","mean","std"], fill_value = 0)
```

```python
pd.pivot_table(df, index=["Account Manager","Contact"], values=["Sale Price","Licenses"], columns=["Product"], aggfunc=('sum','mean','std'), fill_value = 0)
```
